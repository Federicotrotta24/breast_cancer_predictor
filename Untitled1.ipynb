# Diagnosing Breast Cancer from Histology Features

Here we will use the "Breast Cancer Wisconsion" dataset to train a machine learning model to prdict a breast cancer sample is malignat or bening. 

```python
import pandas as pd
import numpy as np
import seaborn as sns
import plotly.express as px
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_repor
```

```python
data = pd.read_csv('/Users/federicotrotta/Downloads/data.csv')
```

## Data Exploration

```python
# Function for Data Exploration
def explore_data(df):
    # Check for missing values
    print("Missing Values:")
    print(df.isnull().sum())

    # Explore the distribution of the target variable
    print("\nTarget Variable Distribution:")
    print(df['diagnosis'].value_counts())

    # Visualize relationships between features and target variable
    # Example: Pairplot for selected columns
    sns.pairplot(df[['diagnosis', 'radius_mean', 'texture_mean', 'area_mean']], hue='diagnosis')
    plt.show()
```

```python
data = data.drop('Unnamed: 32', axis=1)
data = data.drop('id', axis=1)
```

## Data Preprocessing

<b>Here, what we're doing is replacing the diagnosis with 0 or 1 to make the machine learning algorithm work.

```python
data['diagnosis'].replace({'M':1, 'B': 0}, inplace=True)
```

## Train-Test Split

```python
features = data.drop('diagnosis', axis=1) # Esto es la X
target = data['diagnosis'] # Esto es la y 

X_train, X_test, y_train, y_test = train_test_split(features, target, 
test_size=0.2, random_state=42)
```

## Model Selection


```python
model = RandomForestClassifier(random_state=42)
```

Random Forests can capture complex relationships and interactions between features. They consist of an ensemble of decision trees, and each tree learns different aspects of the data. This ensemble approach can be beneficial when dealing with intricate patterns.

## Model Training 

```python

# Train the model
model.fit(X_train, y_train)
```

## Model Evaluation

```python

y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy}')
print(classification_report(y_test, y_pred))
```

## Let use a Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
conf_matrix = confusion_matrix(y_test, y_pred)
conf_matrix
```

```python
fig, ax = plt.subplots(figsize=(7.5, 7.5))
ax.matshow(conf_matrix, cmap=plt.cm.Blues, alpha=0.3)
for i in range(conf_matrix.shape[0]):
    for j in range(conf_matrix.shape[1]):
        ax.text(x=j, y=i,s=conf_matrix[i, j], va='center', ha='center', size='xx-large')

plt.xlabel('Predictions', fontsize=18)
plt.ylabel('Actual Values', fontsize=18)
plt.title('Confusion Matrix', fontsize=18)
plt.show()
```

*Components of a Confusion Matrix:*

**True Positive (TP):** Instances where the model correctly predicted the positive class. In this case the values is: **70**

**True Negative (TN)**
Instances where the model correctly predicted the negative class. In this case the values is: **40**

**False Positive (FP):**
Instances where the model predicted the positive class, but the true class was negative.
In this case the values is: **1**

**False Negative (FN):**
Instances where the model predicted the negative class, but the true class was positive.
In this case the values is: **3**
